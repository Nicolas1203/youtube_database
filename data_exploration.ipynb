Compute Covariances
---------

Standardize
--

```python
def center(X):
    return X - np.mean(X, axis=0)

def standardize(X):
    newX = center(X)/np.std(X, axis=0)
    return newX

df = standardize(df)
df.to_csv('./data/data_all_standardize.csv', index=False)
```

```python
cov_matrix = df.cov()
cov_matrix
```

```python
import youtube_dl as yt
import pandas as pd
import time
import vtt2txt
import glob
```

```python
# Define YTDL options
opts = {
    "format": '18',
    "writesubtitles": True,
    "writeautomaticsub": True,
#    "all_sub": True,
    "subtitleslangs": {
        "fr",
    },
    "skip_download": True
}
```

```python
ytdl = yt.YoutubeDL(opts)

df = pd.read_csv("id_videos_commentaires_SANS_ss_titres.csv")
print(df.head())
url_list = df['video_id'].apply(lambda x: 'https://www.youtube.com/watch?v=' + str(x)).tolist()
nb_unavailable = 0
time_start = time.time()
for url in url_list:
    try:
        ytdl.download([url])
    except Exception as e:
        print(f"{url} video is unavailable, skipping it")
        nb_unavailable += 1
print(f"Number of skipped videos : {nb_unavailable}")
print(f"Processing time {time.time() - time_start}")
```

```python
import pandas as pd

df = pd.read_csv("id_videos_commentaires_SANS_ss_titres.csv")
df['auto_sub'] = df['video_id'].apply(lambda x: "")
df
```

```python
txt_files = glob.glob('/home/nicolas/Documents/thesis/youtube_database/vtt_files/stats_for_bilel/autofr/*.txt')

for i in range(len(df)):
    line = df.iloc[i]
    for file_name in txt_files:
        if line['video_id'] in file_name:
            with open(file_name) as f:
                df.loc[i]['auto_sub'] = f.read()
                f.close()
df
```
