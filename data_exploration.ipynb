Data preprocessing
=====

NaN filtering
----

```python
# ratio of NaN
nb_na = df.isna().sum() / len(df)
print("Ratio of NaN")
print(nb_na.sort_values())
```

```python
print("Global ratio of not NaN : ", df.count().sum() / df.size)
```

```python
# Removes columns where there is more than 34% of NaN
col_to_keep = nb_na[nb_na < 0.34].index.tolist()
print("Number of kept columns")
print(len(col_to_keep))
df = df[col_to_keep]
```

```python
df = df.fillna(method='ffill')
```

Compute Covariances
---------

Standardize
--

```python
def center(X):
    return X - np.mean(X, axis=0)

def standardize(X):
    newX = center(X)/np.std(X, axis=0)
    return newX

df = standardize(df)
df.to_csv('./data/data_all_standardize.csv', index=False)
```

```python
cov_matrix = df.cov()
cov_matrix
```

```python


fig, ax = plt.subplots()


ax.matshow(cov_matrix, cmap=plt.cm.Blues)
plt.show
```
